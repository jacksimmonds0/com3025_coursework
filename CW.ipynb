```python
import pandas as pd
import gzip

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df = getDF('input/reviews_Electronics.json.gz')

df1 = getDF('input/meta_Electronics.json.gz')
```

```python
print(df["product"])
```

```python
df1 = df1.drop(columns=['imUrl', 'brand', 'salesRank'])
```

```python
df = df.drop(columns=['unixReviewTime'])
```

```python
combined = df1.join(df.set_index('asin'), on ='asin')
```

```python
combined.head()
```

```python

## remove any reviews that are null
combined_filtered = combined.dropna(subset=['reviewText'])

## process a subset, for now
combined_filtered_subset = combined_filtered.head(1000)

##duplicate column to preserve original
combined_filtered_subset['reviewTextProcessed'] = combined_filtered_subset['reviewText']

combined_filtered_subset.head()
```

```python
import re

REPLACE_NO_SPACE = re.compile("[.;:!\'?,\"()\[\]]")
REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)")

def preprocess_reviews(reviews):
    reviews = [REPLACE_NO_SPACE.sub("", line.lower()) for line in reviews]
    reviews = [REPLACE_WITH_SPACE.sub(" ", line) for line in reviews]
    
    return reviews

reviews_clean = preprocess_reviews(combined_filtered_subset['reviewTextProcessed'].tolist())
```

```python
from nltk.corpus import stopwords

english_stop_words = stopwords.words('english')
def remove_stop_words(corpus):
    removed_stop_words = []
    for review in corpus:
        removed_stop_words.append(
            ' '.join([word for word in review.split() 
                      if word not in english_stop_words])
        )
    return removed_stop_words

no_stop_words = remove_stop_words(reviews_clean)
```

```python
from nltk.stem.porter import PorterStemmer

def get_stemmed_text(corpus):
    stemmer = PorterStemmer()
    return [' '.join([stemmer.stem(word) for word in review.split()]) for review in corpus]

stemmed_reviews = get_stemmed_text(no_stop_words)
print(stemmed_reviews[1])
```

```python
def get_lemmatized_text(corpus):
    from nltk.stem import WordNetLemmatizer
    lemmatizer = WordNetLemmatizer()
    return [' '.join([lemmatizer.lemmatize(word) for word in review.split()]) for review in corpus]

lemmatized_reviews = get_lemmatized_text(stemmed_reviews)
print(lemmatized_reviews)
```
